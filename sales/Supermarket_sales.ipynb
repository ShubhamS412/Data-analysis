# Finding market share

```python
loc_list = d.groupby('Location')
```

```python
loc = [x for x,y in loc_list]
loc
```

```python
d.groupby('Location').sum()['Total']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(loc,d.groupby('Location').sum()['Total'])
plt.show()
```

```python
plt.pie(d.groupby('Location').sum()['Total'],labels=loc,autopct='%1.1f%%')
plt.show()
```

# Classifying Shoppers

```python
d.groupby(['Location','Gender']).count()
```

```python
lo = d.groupby(['Location','Gender']).count()['Invoice ID']
uskd_sales =lo.unstack(level=0)
```

```python
uskd_sales.plot(kind='bar',figsize = (16,9))
```

```python
lo1 = d.groupby(['Gender','Location']).count()['Invoice ID']
uskd_sales1 =lo1.unstack(level=0)
```

```python
uskd_sales1.plot(kind='bar')
```

# Analysing memberships and ratings

```python
members = d.groupby(['Member','Location']).count()['Invoice ID']
members.unstack(level=0).plot(kind='bar',figsize = (16,9))
plt.show()
```

```python
rating = d.groupby('Location').mean()['Rating']
rating
```

```python
rating.plot(kind='bar')
plt.show()
```
